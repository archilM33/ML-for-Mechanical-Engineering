1. Write a NumPy program to reverse an array

```python
import numpy as np
stringg = input("enter the elements of array separated by spaces: ")
array = np.fromstring(stringg ,dtype = float, sep=" ")
print(array)
i = 0
length = len(array)
for i in range(len(array)//2):
    tempp = array[i]
    array[i] = array[length-i-1]
    array[length-i-1] = tempp
print(array)

```

2. Write a NumPy Program for finding the determinant of the 3D input matrix

```python
first = input("Enter the first size ")
sec = input("Enter the second size ")
third = input("Enter the third size " )
matrix = np.zeros((int(first),int(sec),int(third)))
for i in range(int(first)):
    for j in range(int(sec)):
        for k in range(int(third)):
            matrix[i][j][k] = input("Enter the element ")
print(matrix)
print("The determinant of the matrix is ")
print(np.linalg.det(matrix))
```

3.Write a numpy program for finding the traces of determinant

```python
row = input("Enter the number of rows ")
col = input("Enter the number of columns ")
matrix = np.zeros((int(row),int(col)))
for i in range(int(row)):
    for j in range(int(col)):
        matrix[i][j] = input("Enter the element ")
if i!=j:
    print("The matrix is not a square matrix")
else:
    print(matrix)
print("The trace of the matrix is ")
print(np.trace(matrix))
```

4. Create a Fibonacci series and find the sum of first 100 terms

```python
def fibonacci_sum(n):

  f_list = [0, 1]
  if n <= 2:
      return sum(f_list[:n])

  for i in range(2, n):
      next_f = f_list[i - 1] + f_list[i - 2]
      f_list.append(next_f)

  return sum(f_list)


terms = 100
result = fibonacci_sum(terms)
print("The sum of the first 100 Fibonacci numbers is: ",fibonacci_sum(terms))
```
